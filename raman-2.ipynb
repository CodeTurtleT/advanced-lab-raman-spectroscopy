```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter, find_peaks
from scipy.optimize import curve_fit

messdaten_auf_steg = np.genfromtxt('auf_Steg.txt', delimiter='\t',
    converters={
        0: lambda s: float(s.replace(',', '.')),
        1: lambda s: float(s.replace(',', '.'))
    }
)

wellenzahl_cm = messdaten_auf_steg[:,0]
amplitude = messdaten_auf_steg[:,1]

amplitude_smooth = savgol_filter(amplitude, window_length=11, polyorder=3)

peaks, _ = find_peaks(amplitude_smooth, prominence=0.02)

def lorentz(x, A, x0, gamma):
    return A / (1 + ((x - x0)/gamma)**2)

mask_D = (wellenzahl_cm > 1200) & (wellenzahl_cm < 1450)
mask_G = (wellenzahl_cm > 1500) & (wellenzahl_cm < 1700)

x_D = wellenzahl_cm[mask_D]
y_D = amplitude[mask_D]

x_G = wellenzahl_cm[mask_G]
y_G = amplitude[mask_G]

popt_D, _ = curve_fit(lorentz, x_D, y_D, p0=[max(y_D), 1350, 50])
popt_G, _ = curve_fit(lorentz, x_G, y_G, p0=[max(y_G), 1580, 30])


A_D, x0_D, gamma_D = popt_D
A_G, x0_G, gamma_G = popt_G

print("D-Peak Position:", x0_D)
print("G-Peak Position:", x0_G)

ID_IG = A_D / A_G
print("I_D / I_G =", ID_IG)

plt.figure(figsize=(10,5))

plt.plot(wellenzahl_cm, amplitude, label="Rohdaten", alpha=0.5)
plt.plot(wellenzahl_cm, amplitude_smooth, label="geglättet")

plt.plot(x_D, lorentz(x_D, *popt_D), label="D-Fit")
plt.plot(x_G, lorentz(x_G, *popt_G), label="G-Fit")

plt.legend()
plt.xlim(1000, 1700)
plt.xlabel("Raman shift [cm⁻¹]")
plt.ylabel("Intensität")
plt.grid()
plt.show()
```
